```python
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
def generate_save_path(mode,num,wc, scope):
    save_path = './'+mode+'/result-'+wc+'-'+str(scope)+'/tgModels'+str(num)+'/'
    return save_path
```

```python
def generate_clone_path(mode,num,wc, scope):
    save_path = './'+mode+'/clone-'+wc+'-'+str(scope)+'/tgModels'+str(num)+'/'
    return save_path
```

```python
def clone_model(mode,wc,scope):
    for num in range(308):
        save_path = generate_save_path(mode,num,wc,scope)
        save_path_clone = generate_clone_path(mode,num,wc,scope)
        # will return the path to the cloned predictor, identical to save_path_clone
        predictor = TabularPredictor.load(save_path)
        path_clone = predictor.clone_for_deployment(path=save_path_clone)
#     generate_save_path('x',1,'high',1)
```
